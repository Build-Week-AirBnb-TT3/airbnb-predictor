### load in models and apply knn

```python
tfidf = pickle.load(open('tfidf.pkl', 'rb'))
```

```python
dtm = tfidf.fit_transform(df1['amenities'])
dtm = pd.DataFrame(dtm.todense(), columns=tfidf.get_feature_names())
```

```python
# knn 
knn = NearestNeighbors(n_neighbors=10, metric='cosine')
knn.fit(dtm)
```

```python
def recommender(text):
  x = pd.DataFrame(columns=df1.columns)
  input_features = tfidf.transform(text)
  for i in knn.kneighbors(input_features, n_neighbors=5, return_distance=False)[0]:
    x = x.append(df1.iloc[[i]])
  return x
```
