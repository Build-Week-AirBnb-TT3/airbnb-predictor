```python
def baseline_model():
    model=Sequential()
    model.add(Dense(174, input_dim=174, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', 
                  optimizer='adam',
                  metrics=['mean_absolute_error'])
    return model
```

```python
X_enc_train, X_enc_test, y_enc_train, y_enc_test = train_test_split(
 X_enc, y_enc, test_size=0.2, random_state=42)
```

```python
stop = EarlyStopping(monitor='mean_absolute_error',
                    min_delta=0.01,
                    patience=3)
```

```python
def build(hp):
    hp_opt = hp.Choice('opt', values=['Adam', 'Nadam'])
    hp_lr = hp.Float('lr', min_value=.0005, max_value=.0015)
    hp_act_funct_a = hp.Choice('act_funct', values=['relu', 'tanh'])
#     hp_act_funct_b = hp.Choice('act_funct', values=['ReLU', 'ThresholdedReLU', 'PReLU'])
    hp_units = hp.Int('units', min_value=32, max_value=512, step=32)
    hp_layers = hp.Int('layers', min_value=1, max_value=20, step=1)
    
    model = Sequential()
    model.add(layers.Dense(hp_units,
                            activation=hp_act_funct_a,
                            input_dim=174))

    for i in range(hp_layers):
        model.add(layers.Dense(units=hp_units,
                               activation=hp_act_funct_a))
    model.add(layers.Dense(1))
    
    
    model.compile(
        optimizer=hp_opt,
        loss='mean_squared_error',
        metrics=['mean_absolute_error'])
    
    return model
```

```python
keras.backend.clear_session()
```

```python
tuner = kt.BayesianOptimization(build,
                               objective = 'mean_absolute_error',
                               max_trials=100,
                               overwrite=True)
```

```python
tuner.search(X_enc_train, y_enc_train,
            validation_data=(X_enc_test, y_enc_test),
            epochs=100,
            callbacks=[stop], 
            verbose=1,
            batch_size=10)
```

```python
tuner2 = kt.BayesianOptimization(build,
                               objective = 'mean_absolute_error',
                               max_trials=5)
```

```python
tuner2.search(X_enc_train, y_enc_train,
            validation_data=(X_enc_test, y_enc_test),
            epochs=1000,
            callbacks=[stop], 
            verbose=1,
            batch_size=10)
```
